DROP UNNAMED COLOUM

```python
df = df.drop("Unnamed: 0", axis = 1)
print(df.head())
```

```python
df.head()
```

```python
# change weekly study hours
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("05-Oct", "05-10")
```

```python
df.head()
```

gender distribution

```python
plt.figure(figsize = (4,5))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

From the above chart we have analysed that:
the number of females in the data is more than the number of males

Ethnic group distribution

```python
plt.figure(figsize = (4,5))
ax = sns.countplot(data = df, x = "EthnicGroup")
ax.bar_label(ax.containers[0])
plt.title("Distribution on Ethnic Group")
plt.show()
```

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()

l = ["group A","group B","group C","group D","group E"]
mylist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]
plt.pie(mylist, labels = l, autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize = (6,4)) 
sns.heatmap(gb, annot = True)
plt.title("Relationship b/w Parent's Education and student's Score")
plt.show()
```

from the above chart we have concluded that the education of the parents havea good impacts on their scores

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb1)
```

```python
plt.figure(figsize = (6,4)) 
sns.heatmap(gb1, annot = True)
plt.title("Relationship b/w Parent's Marital Status and student's Score")
plt.show()
```

from the above chart we have concluded that there is no impact on the student's score due to their parent's marital status.
